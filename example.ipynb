# Otsu threshold implementation using 1st order statistical moments

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import ipywidgets
from otsu_segmentation import otsu
```

```python
for img_id in [1,2,3]:
    img = cv2.imread(f"demo_images/demo{img_id}.png",cv2.IMREAD_GRAYSCALE)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 4))

    plt.subplot(121)
    cv2_otsu_threshold,img1 = cv2.threshold(img,0,255,cv2.THRESH_OTSU)
    plt.imshow(img1, cmap='gray', vmin=0, vmax=1)
    plt.colorbar(fraction=0.035, pad=0.04,ticks=[0, 1])
    plt.xlabel("x [px]")
    plt.ylabel("y [px]")
    plt.title('cv2 implementation')

    plt.subplot(122)
    otsu_threshold = otsu(img/np.max(img))
    binary_image = (img>=otsu_threshold).astype(int)
    plt.imshow(binary_image, cmap='gray', vmin=0, vmax=1)
    plt.colorbar(fraction=0.035, pad=0.04,ticks=[0, 1])
    plt.xlabel("x [px]")
    plt.ylabel("y [px]")
    plt.title('personal implementation')
    plt.rcParams.update({'font.size': 20})
    plt.show()

    print(f"cv2 implementation : otsu_thr = {int(cv2_otsu_threshold)}")
    print(f"personal implementation : otsu_thr = {otsu_threshold}")
    print("\n")
```

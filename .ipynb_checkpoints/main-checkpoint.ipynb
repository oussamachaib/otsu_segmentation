# Otsu threshold implementation

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import ipywidgets
from otsu_segmentation import otsu
```

```python
img = cv2.imread("demo.png",cv2.IMREAD_GRAYSCALE)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 5))

plt.subplot(121)
cv2_otsu_threshold,img1 = cv2.threshold(img,0,255,cv2.THRESH_OTSU)
plt.imshow(img1, cmap='gray', vmin=0, vmax=1)
plt.colorbar(fraction=0.035, pad=0.04,ticks=[0, 1])
plt.xlabel("x [px]")
plt.ylabel("y [px]")
plt.title('cv2 implementation')

plt.subplot(122)
otsu_threshold = otsu(img/np.max(img))
binary_image = (img>=otsu_threshold).astype(int)
plt.imshow(binary_image, cmap='gray', vmin=0, vmax=1)
plt.colorbar(fraction=0.035, pad=0.04,ticks=[0, 1])
plt.xlabel("x [px]")
plt.ylabel("y [px]")
plt.title('personal implementation')
plt.rcParams.update({'font.size': 20})
plt.show()
```

# Note to self
As expected, the results do not match perfectly, think the difference stems from the implementation of cv2 which applies a 5x5 gaussian filter before computing the threshold for obscure reasons (in addition to being needlessly lengthy : https://docs.opencv.org/4.x/d7/d4d/tutorial_py_thresholding.html). Another reason could be normalization from the grayscale dynamic range to a [0,1] domain.
\
\
Moral of the story: the in-house code is more reliable.
